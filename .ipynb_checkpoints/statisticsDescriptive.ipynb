Probability  Density Function
Density Extimation

```python
from numpy.random import normal
```

```python
sample=normal(loc=50,scale=5,size=1000)
plt.hist(sample)
plt.show()
```

```python
sample_mean=sample.mean()
sample_std=sample.std()
```

```python
from scipy.stats import norm
```

```python
dist=norm(sample_mean,sample_std)
```

```python
values=np.linspace(sample.min(),sample.max(),100)
```

```python
probabilityDensity=[dist.pdf(value) for value in values]
plt.hist(sample,bins=10,density=True)
plt.plot(values,probabilityDensity)
plt.show()
```

```python
import seaborn as sns
```

```python
sns.displot(sample,kind='hist')
```

Non Parameter Density Estimation 
kernel Density Estimation

```python
sample1=normal(loc=20,scale=5,size=300)
sample2=normal(loc=40,scale=5,size=700)
sample=np.hstack((sample1,sample2))
```

```python
plt.hist(sample,bins=50)
plt.show()
```

```python
from sklearn.neighbors import KernelDensity
```

```python
model=KernelDensity(bandwidth=3,kernel='gaussian')
sample=sample.reshape(len(sample),1)
model.fit(sample)
```

```python
values=np.linspace(sample.min(),sample.max(),100)
values=values.reshape(len(values),1)
```

```python
probability=model.score_samples(values)
#give the value in log form so covert log to normal we are use the exponential form
probability=np.exp(probability)
```

```python
plt.hist(sample,bins=50,density=True)
plt.plot(values[:],probability)
plt.show()
```

```python
sns.displot(sample.reshape(1000),kind='kde',bw_adjust=1)
```

```python
sampleData=np.random.randint(low=1,high=7,size=1000)
series=pd.Series(sampleData)
p=(series.value_counts()/(series.value_counts().sum())).sort_index()
p.plot(kind='bar')
```

```python
cdf=np.cumsum(p)
pd.Series(cdf).plot(kind='bar')
```
