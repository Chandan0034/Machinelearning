# 3D Surface Plots

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
xx,yy=np.meshgrid(x,y)
xx.shape
```

```python
z=xx**2+yy**2
z.shape
```

```python
fig=plt.figure(figsize=(12,8))
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
fig.show()
```

```python
z=np.sin(xx)+np.cos(yy)
fig=plt.figure(figsize=(12,8))
ax=plt.subplot(projection='3d')
a=ax.plot_surface(xx,yy,z,cmap='viridis')
plt.colorbar(a)
fig.show()
```

```python
fig=plt.figure(figsize=(12,8))
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
fig.show()
```

Contour Plots

```python
fig=plt.figure(figsize=(12,8))
ax=plt.subplot(projection='3d')
p=ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
fig.show()
```

```python
fig=plt.figure(figsize=(15,8))
ax=plt.subplot(projection='3d')
p=ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
fig.show()
```

```python
df=pd.read_csv('C:\\Users\\91930\\100dayML\\IPL_Ball_by_Ball_2008_2022.csv')
df.head()
```

```python
df['ballnumber'].unique()
```

```python
temp_df=df[(df['batsman_run']==6) & (df['ballnumber'].isin([1,2,3,4,5,6]))]
temp_df.head()
```

```python
grid=temp_df.pivot_table(index='overs',values='batsman_run',columns='ballnumber',aggfunc='count')
grid
```

```python
plt.figure(figsize=(12,15))
plt.imshow(grid)
plt.yticks(np.arange(0,20),list(range(1,21)))
plt.xticks(np.arange(0,6),list(range(1,7)))
plt.colorbar()
```
