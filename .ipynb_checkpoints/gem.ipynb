```python
import numpy as np
from selenium import webdriver
import pandas as pd
from bs4 import BeautifulSoup
# import requests
```

```python
name=[]
badge=[]
price=[]
description=[]
image=[]
```

```python
def collectDataFromGem(start,end):
        for i in range(start,end):
                gemUrl="https://mkp.gem.gov.in/computers-0806nb/search#/?q[]=laptop&page={}".format(i)
                driver=webdriver.Edge()
                driver.get(gemUrl)
                driver.implicitly_wait(10)
                html_content=driver.page_source
                driver.quit()
                soup=BeautifulSoup(html_content,'html.parser')
                laptopList=soup.find_all('div',class_='variant-wrapper btn-add_to_cart rle-add_to_cart bre- price-dekho-enabled product-init')
                print(len(laptopList))
                for i in laptopList:
                        try:
                                name.append(i.find('span',class_='variant-title').text)
                        except:
                                name.append(np.nan)
                        try:
                                badge.append(i.find('span',class_='badge').text)
                        except:
                                badge.append(np.nan)
                        try:
                                price.append(i.find('span',class_='variant-final-price').text.split('\n')[1])
                        except:
                                price.append(np.nan)
                        try:
                                image.append((i.find('img')).get('src'))
                        except:
                                image.append(np.nan)
collectDataFromGem(10,20)
d={"Name":name,"Price":price,"Rating":badge,"Image":image}
df=pd.DataFrame(d)
```

```python
df.to_csv('gemData.csv')
df
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup

def search_amazon(query):
    driver = webdriver.Edge()
    driver.get("https://www.amazon.in/")
    # Find the search input element and send the query
    search_box = driver.find_element("id", "twotabsearchtextbox")
    search_box.send_keys(query)
    search_box.send_keys(Keys.RETURN)
    driver.implicitly_wait(20)
    html_content = driver.page_source
    driver.quit()
    return html_content
def extract_laptops(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')
    laptops = soup.find_all('div', class_='a-section a-spacing-small a-spacing-top-small')
    for laptop in laptops:
        try:
            title_element = laptop.find('span', class_='a-size-medium a-color-base a-text-normal')
        except:
            title_element=np.nan
        try:
            price_element = laptop.find('span', class_='a-price-whole')
        except:
            price_element=np.nan
        try:
            rating_element=laptop.find('span',class_='a-icon-alt')
        except:
            rating_element=np.nan
        try:
            image_element=laptop.find('img').get('src')
        except:
            image_element=np.nan
        
        if title_element and price_element and rating_element and image_element:
            title.append(title_element.text.strip())
            price.append(price_element.text.strip())
            rating.append(rating_element.text)
            image.append(image_element)
# search_query = "hp Intel Core i7 15.6 Inch Laptop"
# html_content = search_amazon(search_query)
# extract_laptops(html_content)
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException, ElementClickInterceptedException
import time

def get_flipkart_price(product_name):
    # Path to GeckoDriver
    PATH = "/Users/sauravjain/Downloads/geckodriver"
    option=webdriver.FirefoxOptions()
    option.add_argument('--headless')
    driver = webdriver.Firefox(options=option)

    try:
        # Open Flipkart
        driver.get("https://www.flipkart.com")

        # Close the login pop-up if it appears
        try:
            close_button = driver.find_element(By.CSS_SELECTOR, "button._2KpZ6l._2doB4z")
            close_button.click()
        except NoSuchElementException:
            pass  # No pop-up appeared

        # Find the search bar, enter the product name, and submit
        search_bar = driver.find_element(By.NAME, "q")
        search_bar.send_keys(product_name)
        search_bar.send_keys(Keys.RETURN)

        # Wait for the page to load
        time.sleep(2)

        # Find and print the price of the first product in search results
        price = driver.find_element(By.CSS_SELECTOR, '._1_WHN1').text
        rating = driver.find_element(By.CSS_SELECTOR, '._1lRcqv').text
        print(f"Price of '{product_name}': {price}")
        print(f"Rating of '{product_name}': {rating}")

    except (NoSuchElementException, ElementClickInterceptedException) as e:
        print("Error occurred.")
        print(e)
    finally:
        # Close the browser
        driver.quit()
  
# Example usage
get_flipkart_price("APPLE iPhone 13 (Gold, 128 GB)")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
import time

def get_flipkart_product_info(product_name):
    # Specify the path to your WebDriver (replace with the path to your chromedriver or geckodriver)
    chromedriver_path = "/path/to/chromedriver"
    chrome_option=webdriver.ChromeOptions()
    chrome_option.add_argument('--headless')
    # Initialize the WebDriver (use Chrome or Firefox based on your preference)
    driver = webdriver.Chrome(options=chrome_option)

    try:
        # Open Flipkart
        driver.get("https://www.flipkart.com")

        # Close the login pop-up if it appears
        try:
            close_button = driver.find_element(By.CSS_SELECTOR, "button._2KpZ6l._2doB4z")
            close_button.click()
        except NoSuchElementException:
            pass  # No pop-up appeared

        # Find the search bar, enter the product name, and submit
        search_bar = driver.find_element(By.NAME, "q")
        search_bar.send_keys(product_name)
        search_bar.send_keys(Keys.RETURN)

        # Wait for the page to load
        time.sleep(2)

        # Find and print the price and rating of the first product in search results
        try:
            price = driver.find_element(By.CSS_SELECTOR, '._30jeq3._16Jk6d').text
            rating = driver.find_element(By.CSS_SELECTOR, '._1lRcqv').text
            print(f"Price of '{product_name}': {price}")
            print(f"Rating of '{product_name}': {rating}")
        except NoSuchElementException:
            print(f"Product '{product_name}' not found.")

    except Exception as e:
        print(f"An error occurred: {e}")
    finally:
        # Close the browser
        driver.quit()

# Example usage
get_flipkart_product_info("iphone 13")
```
