```python
def sanity_check(data_type):
    def outerWrapper(func):
        def inner_wrapper(*args):
            print(args)
            if type(args[0])==data_type:
                func(*args)
            else:
                raise TypeError("ye data type nai chalega")
        return inner_wrapper
    return outerWrapper
@sanity_check(int)
def square(num,b):
    print(num**b)
square(5,3)
```

```python
import pickle
class Person:
    def __init__(self):
        pass
    def square(self,a,b):
        print(a**b)
person=Person()
with open('sample.pkl','wb') as f:
    pickle.dump(person,f)
```

```python
with open('sample.pkl','rb') as f:
    person=pickle.load(f)
    person.square(5,6)
```
